# Data Processing

```python
# Imports
import numpy as np
import mne
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
import pickle
```

```python
def resample_and_save_eeg(input_fif_path, output_dir, target_sfreq=200):
    # Load the Epochs data
    epochs = mne.read_epochs(input_fif_path, preload=True)
    
    # Resample
    epochs.resample(sfreq=target_sfreq)
    
    # Generate output path
    base_name = os.path.splitext(os.path.basename(input_fif_path))[0]
    output_filename = f"{base_name}_downsampled_{int(target_sfreq)}.fif"
    output_fif_path = os.path.join(output_dir, output_filename)

    os.makedirs(output_dir, exist_ok=True)
    epochs.save(output_fif_path, overwrite=True)
    print(f"✔️ Saved: {output_fif_path}")
```

```python
# Paths
input_root = "data/PreprocessedEEGData"
output_root = "data/PostprocessedEEGData"

# Loop through the IDs and versions
for subject_id in range(301, 332):  # 301 to 331 inclusive
    for version in ['A', 'B', 'C']:
        filename = f"{subject_id}{version}_FG_preprocessed-epo.fif"
        input_path = os.path.join(input_root, filename)
        
        if os.path.exists(input_path):
            resample_and_save_eeg(input_path, output_root, target_sfreq=200)
        else:
            print(f"⚠️ File not found: {input_path}")
```
